```python
import seaborn as sns 
import pandas as pd 
import numpy as np 

%matplotlib inline
```

```python
x = np.array([1.2,3.2,4.5,6.5,4.9,7.7])
y = np.array([5.4,5.1,3.2,4.3,4.4,2.1])

df = pd.DataFrame({'x' : x,'y' : y})

sns.regplot('x','y',data=df,scatter=True)
```

```python
df_iris = pd.read_csv('iris.csv',dtype={'Species' : 'category'})
df_iris.head(5)
```

```python
sns.pairplot(df_iris.drop(['Id','Species'],axis=1))
```

```python
sns.pairplot(df_iris.drop(['Id'],axis=1),hue='Species')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
df_iris['Species'].unique()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
colors = {'Iris-setosa' : 'red','Iris-versicolor' : 'green', 'Iris-virginica' : 'blue'}
ax.scatter(df_iris['SepalLengthCm'],df_iris['SepalWidthCm'],df_iris['PetalLengthCm'],c=df_iris['Species'].apply(lambda x : colors[x]))
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalLengthCm')
```
